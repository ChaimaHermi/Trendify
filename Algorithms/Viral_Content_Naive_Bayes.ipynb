```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=3000)
X_hashtags = tfidf.fit_transform(df["hashtags"].fillna(""))
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse_output=True, handle_unknown="ignore")
X_content = ohe.fit_transform(df[["content_type"]])
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_sentiment = scaler.fit_transform(df[["sentiment_score"]])
```

```python
X = hstack([
    X_hashtags,
    X_content,
    X_sentiment
]).tocsr()   # 👈 TRÈS IMPORTANT

```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
import numpy as np
```

```python
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

model = MultinomialNB(alpha=1.0)

scores = []

for train_idx, test_idx in skf.split(X, y):
    X_train, X_test = X[train_idx], X[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    acc = accuracy_score(y_test, y_pred)
    scores.append(acc)
```

```python
print("Accuracy par fold :", scores)
print("Accuracy moyenne :", np.mean(scores))
```

```python
# Texte
X_hashtags = df["hashtags"]

# Catégorielles
X_content_type = df[["content_type"]]
X_topic = df[["topic"]]
X_platform = df[["platform"]]

# Numériques
X_numeric = df[["views", "comments", "shares"]]

# Numérique déjà utilisé
X_sentiment = df[["sentiment_score"]]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(
    max_features=3000,
    ngram_range=(1, 2)
)

X_hashtags_tfidf = tfidf.fit_transform(X_hashtags)
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(
    sparse_output=True,
    handle_unknown="ignore"
)

X_cat = df[["content_type", "topic", "platform"]]
X_cat_ohe = ohe.fit_transform(X_cat)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X_numeric_scaled = scaler.fit_transform(X_numeric)
X_sentiment_scaled = scaler.fit_transform(X_sentiment)
```

```python
from scipy.sparse import csr_matrix

X_numeric_sparse = csr_matrix(X_numeric_scaled)
X_sentiment_sparse = csr_matrix(X_sentiment_scaled)
```

```python
from scipy.sparse import hstack

X = hstack([
    X_hashtags_tfidf,
    X_cat_ohe,
    X_numeric_sparse,
    X_sentiment_sparse
]).tocsr()
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
import numpy as np

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

model = MultinomialNB(alpha=0.5)

scores = []

for train_idx, test_idx in skf.split(X, y):
    X_train = X[train_idx]
    X_test  = X[test_idx]

    y_train = y.iloc[train_idx]
    y_test  = y.iloc[test_idx]

    model.fit(X_train, y_train)
    scores.append(accuracy_score(y_test, model.predict(X_test)))

print("Accuracy moyenne :", np.mean(scores))
```
